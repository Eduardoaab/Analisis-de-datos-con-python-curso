## Merge

```python
#Hace la union de las dos tablas y si no encuentra un valor le pone NaN 
pd.merge(df1,df2, on= "Producto",how= "outer" )
```

```python
#Hace la union de las dos tablas pero si no encontro un valor que no estaba en la primera tabla o en la segunda tabla omite ese producto
pd.merge(df1,df2, on= "Producto",how= "inner" )
```

```python
#el de la izquierda con el de la derecha solo si sus valores aparecen en la derecha 
pd.merge(df1,df2, on= "Producto",how= "right" )
```
