```python
"""
*********************************************
DICCIONARIOS
*********************************************
Los diccionarios son un conjunto de claves (keys) asociados a valores, 
pero que no est ordenado. mediante la clave podemos acceder al vaolor.  
"""

#%%
#los diccionarios se declaran con llaves
#las llaves se definen con dos puntos y pueden ser caulquer variable visto e incluso listas
#La forma de presentar los datos debe ser entendible
inventario = {
     "Madalenas":3,
     "Donas":1,
     "Roles":7
        }

print (type(inventario))

print (inventario)
```

```python
#%%
# Me imprime las llaves que tengo
# No me regresa una lista perse
print(type(inventario.keys()))
```

```python
#%%
#Convertir las llaves de ese diccionario en una lista

list_keys = list(inventario.keys())

print(list_keys[2])
```

```python
#%%
#Para acceder a los valores del diccionario
list_values= list(inventario.values())

print (list_values[2])
```

```python
#%%
#Para acceder a los valores directamente del diccionario
print (inventario ["Madalenas"])
```

```python
#%%

print ("Panques" in inventario)

print ("Donas" in inventario)
```

```python
#%%
#Traer una valor del diccionario y eliminarlo 

pedido  = inventario.pop("Donas")

print ("Tenemos {} paquete de donas".format(pedido))

print (inventario)
```

### Condicional IF

```python
#condicional IF
#Si (condicin)entonces:
# lo que queramos 
#Cualquier otro caso:
#Lo que queramos
edad =17

if edad>18:
    print (" si es mayor de edad")
else:
    print ("no es mayor de edad")
```

```python
# if anidado 
edad =19
pago= False
if edad>18:
    print (" si es mayor de edad")
    if pago == True:
        print ("Entonces eres mayor y pagaste")
    else: 
        print ("Eres mayor y no pagaste")
else:
    print ("no es mayor de edad")
```
